```python
train = train.dropna()
keepers = ["NumberOfTime30-59DaysPastDueNotWorse","NumberOfOpenCreditLinesAndLoans", "NumberOfDependents","age"]

#train_x = train.loc[:,keepers]
train_x = train.loc[:,train.columns[train.columns != "SeriousDlqin2yrs"]]
train_y = train.loc[:,["SeriousDlqin2yrs"]]

# permute = np.random.permutation(len(train_x))

positive_index = (train_y.to_numpy() == 1).flatten()
negative_index = positive_index == False



train_positive = train_x.to_numpy()[positive_index]
train_negative = train_x.to_numpy()[negative_index]
```

```python

def plot_features(f1,f2):
    plt.figure()
    pos_x = train_positive[:,train_x.columns == f1]
    pos_y = train_positive[:,train_x.columns == f2]

    neg_x = train_negative[:,train_x.columns == f1]
    neg_y = train_negative[:,train_x.columns == f2]

    plt.plot(pos_x, pos_y,"or",label="Positive training examples")
    plt.plot(neg_x, neg_y,"ob",label="Negative training examples")

    plt.xlabel(f1)
    plt.ylabel(f2)
    plt.legend(loc='best')
    
plot_features("NumberOfTime30-59DaysPastDueNotWorse","age")

plot_features("NumberOfTime30-59DaysPastDueNotWorse","NumberOfOpenCreditLinesAndLoans")
```

```python


x = []
y = []
n = len(train_x) * 2
for i in range(n):
    roll = np.random.rand()
    if roll > 0.5:
        index = int(np.random.rand() * len(train_positive))
        datum = train_positive[index]
        x.append(datum)
        y.append(1)
    else:
        index = int(np.random.rand() * len(train_negative))
        datum = train_negative[index]
        x.append(datum)
        y.append(0)

permute = np.random.permutation(len(x))

x = np.array(x)
y = np.array(y)

x[permute]
x[permute]

val_percent = 0.2

index = math.floor(len(x)*val_percent)
val_x = x[0:index]
train_x = x[index:]
val_y = y[:index]
train_y = y[index:]
```

```python
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from imblearn.ensemble import BalancedBaggingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, average_precision_score
import math
```

```python
model = BaggingClassifier(n_estimators=100)

model.fit(train_x,train_y)
```

```python
yhat = model.predict(val_x)

y = val_y



print("Accuracy: ",accuracy_score(y,np.round(yhat)))
print("AUC: ", roc_auc_score(y,yhat))
print("PR: ", average_precision_score(y,yhat))
```
