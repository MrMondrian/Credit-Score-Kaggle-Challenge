```python
#train = train.dropna()
train[np.isnan(train)] = 0

permute = np.random.permutation(len(train))


train.reindex(permute)

train_x = train.loc[:,train.columns[train.columns != "SeriousDlqin2yrs"]]
train_y = train.loc[:,["SeriousDlqin2yrs"]]

val_percent = 0.2

index = math.floor(len(train_x)*val_percent)
val_x = train_x[0:index]
train_x = train_x[index:]
val_y = train_y[:index]
train_y = train_y[index:]

keepers = ["NumberOfTime30-59DaysPastDueNotWorse","NumberOfOpenCreditLinesAndLoans", "NumberOfDependents","age","MonthlyIncome"]

#train_x = train.loc[:,keepers]


# permute = np.random.permutation(len(train_x))

positive_index = (train_y.to_numpy() == 1).flatten()
negative_index = positive_index == False



train_positive = train_x.to_numpy()[positive_index]
train_negative = train_x.to_numpy()[negative_index]
```

```python

def plot_features(f1,f2):
    plt.figure()
    pos_x = train_positive[:500,train_x.columns == f1]
    pos_y = train_positive[:500,train_x.columns == f2]

    neg_x = train_negative[:500,train_x.columns == f1]
    neg_y = train_negative[:500,train_x.columns == f2]

    plt.plot(pos_x, pos_y,"or",label="Positive training examples")
    plt.plot(neg_x, neg_y,"ob",label="Negative training examples")

    plt.xlabel(f1)
    plt.ylabel(f2)
    plt.legend(loc='best')
    
plot_features("age","NumberOfTime30-59DaysPastDueNotWorse")

plot_features("NumberOfTime30-59DaysPastDueNotWorse","NumberOfDependents")
```

The above plots show the relationship between NumberOfTime30-59DaysPastDueNotWorse and age as well as NumberOfTime30-59DaysPastDueNotWorse and NumberOfDepents with the label.

# Part 2: Solve the challenge

The crux of my strategy is resmapling the trainging set. Since the data are so imbalanced, to train the model

```python


x = []
y = []
n = len(train_x) * 2
for i in range(n):
    roll = np.random.rand()
    if roll > 0.5:
        index = int(np.random.rand() * len(train_positive))
        datum = train_positive[index]
        x.append(datum)
        y.append(1)
    else:
        index = int(np.random.rand() * len(train_negative))
        datum = train_negative[index]
        x.append(datum)
        y.append(0)

permute = np.random.permutation(len(x))

x = np.array(x)
y = np.array(y)

x = x[permute]
y = y[permute]

val_x = val_x.to_numpy()
val_y = val_y.to_numpy()
```

```python
model = GradientBoostingRegressor(n_estimators=500,loss='absolute_error')

model.fit(x,y)
```

```python
yhat = model.predict(val_x)

y = val_y


print("Accuracy: ",accuracy_score(y,np.round(yhat)))
print("AUC: ", roc_auc_score(y,yhat))
print("PR: ", average_precision_score(y,yhat))
```

# Tell us how you validate your model and why you chose such evaluation technique(s).

I validated my model by withholding 20% of my training data, then evaluating the model's performance on the withheld data. I chose this method, as opposed to an  different method like k-fold cross validation, because my model is very simple and does not have many hyperparamters to tune. More complicated validation techniques like k-fold cross validation are perfect for more complex models like deep neural networks. In my case, I used bagging with decision trees, so the main hyperparameters are tree depth, and number of estimators. I don't need validation to determine the number of estimators, as more is always better so it makes sense to use as many as kaggle will let me get away with. As for tree depth, after evaluating my model using the withheld validation data, I found that a maximum depth of 3 is good, which so happens to be the default.

# What is AUC? Why do you think AUC was used as the evaluation metric for this challenge? What other metrics do you think would also be suitable for this competition? 

AUC stands for area under receiver operator curve. The receiver operator curve (ROC) is a plot that can be generated from model predictions on testing data. To generate this curve, first generate probabilities that the data belong to the positive class. Then define a series of thresholds for what minimum probability constitutes a positive prediction. For each threshold, find the true positive rate and the false positive rate. Plotting these points gives you the ROC. Finally, calculating the area under this curve gives AUC. Better models tend to have high true positive rates and low false positive rates, which results in a high AUC score.

I believe this metric was used for this challenge because of how imblanced the data was. Most of the training data points belong to the negative class, meaning models that disproportionately predict data as negative will achieve a high accuracy. However, this would result in a low true positive rate, and also a low AUC score.

Another metric suitable for this competition would be the area under the precision-recall curve. This metric very similar the AUC, except instead of using the receiver operator curve it uses the precision-recall cruve. The difference is that instead of plotting the true positive rate and false positive rate for each threshold, you calculate the precision and the recall. Recall is another word for true positive rate, and precision is calculated by taking the number of true positives predictions and dividing the total number of positive predictions. This metric perfroms well for imbalanced data and would make sense for this competition.

# Short explanation of what you tried. What worked and what did not work (ie. you might have tried different features/models before the final one). 

The first thing I tried was logistic regression with data normalization and relavent feature selection. The features were selected as the top 5 indepentend features by correrlation with the label. This method achieved good results, approx 70% on AUC. Next I tried bagging with decision trees without normalization or feature selection. This also achieved an AUC around 70%. Next, I tried bagging with logistic regression which didn't make an improvement.

After trying a few methods and not making much progress, I decided to take a very different approach by feeding the data through a dense neural network using tensor flow. I used 5 hidden layers, each with 128 units using relu activation and dropout. Using this I was able to achieve an AUC of 82%. I was happy with this as it is close to the scores seen on the leader board, but I thought I could do better.

Finally, I decided to return to bagging with decision trees. I realized that the issue with my models was that they weren't account for the imbalanced dataset. To account for this, I created a bootstrapped dataset containing an equal number of data points from each class. Finally, I fed this to a model using bagging with 50 decision trees. This achieved an AUC of 99%.
