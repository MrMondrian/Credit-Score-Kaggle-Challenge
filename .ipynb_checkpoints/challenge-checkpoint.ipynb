```python
 
params = [50,100,200,300,500,700,1000]


def auc(model,x,y):
    yhat = model.predict(x)
    yhat = np.clip(yhat,0,1)
    return roc_auc_score(y,yhat)

def validate(params):
    val_scores = []
    for param in params:
        model = HistGradientBoostingRegressor(max_iter=param,scoring=auc)
        scores = cross_val_score(model,x,y,scoring=auc)
        score = sum(scores)/len(scores)
        val_scores.append(score)


    plt.figure()
    plt.plot(params,val_scores,'-')
    plt.xlabel("Max iterations")
    plt.ylabel("AUC")
    plt.show()
validate(params)
```

```python
model = HistGradientBoostingRegressor(max_iter=1000,scoring=auc)
model.fit(x,y)
yhat = model.predict(test.to_numpy())
yhat = np.clip(yhat,0,1)

ids = np.arange(len(yhat)) + 1

answers = pd.DataFrame({"Id":ids,"Probability":yhat})
answers.to_csv("submission6.csv",index=False)
# #np.savetxt("answers.csv", yhat, delimiter=",")
```

# What is AUC? Why do you think AUC was used as the evaluation metric for this challenge? What other metrics do you think would also be suitable for this competition? 

AUC stands for area under receiver operator curve. The receiver operator curve (ROC) is a plot that can be generated from model predictions on testing data. To generate this curve, first generate probabilities that the data belong to the positive class. Then define a series of thresholds for what minimum probability constitutes a positive prediction. For each threshold, find the true positive rate and the false positive rate. Plotting these points gives you the ROC. Finally, calculating the area under this curve gives AUC. Better models tend to have high true positive rates and low false positive rates, which results in a high AUC score.

I believe this metric was used for this challenge because of how imblanced the data was. Most of the training data points belong to the negative class, meaning models that disproportionately predict data as negative will achieve a high accuracy. However, this would result in a low true positive rate, and also a low AUC score.

Another metric suitable for this competition would be the area under the precision-recall curve. This metric very similar the AUC, except instead of using the receiver operator curve it uses the precision-recall cruve. The difference is that instead of plotting the true positive rate and false positive rate for each threshold, you calculate the precision and the recall. Recall is another word for true positive rate, and precision is calculated by taking the number of true positives predictions and dividing the total number of positive predictions. This metric perfroms well for imbalanced data and would make sense for this competition.

# Short explanation of what you tried. What worked and what did not work (ie. you might have tried different features/models before the final one). 

The first thing I tried was logistic regression with data normalization and relavent feature selection. The features were selected as the top 5 independent features by correrlation with the label. This method achieved 40% on AUC. Next I tried bagging with decision trees without normalization or feature selection. This achieved an AUC around 50%. Then, I tried bagging with logistic regression which didn't make an improvement.

I decided to take a very different approach by feeding the data through a dense neural network using tensor flow. I used 2 hidden layers, each with 12 units using relu activation and dropout. Using this I was able to achieve an AUC of 83%.

Finally, I decided to return to ensemble learning. This time, I tried gradient boosting. After using k-fold cross validation to determine the best number of maximum iterations, I found that maximum number of iterations does not have a strong effect on the performance of the model. I re-trained the model with all the training data. This model achieved an AUC of 86%!

# What insight(s) do you have from your model(s)?

One insight I have from my model is ensemble learning is a powerful method for reducing bias and variance. Since none of the input features have a very strong correlation with the label, non-ensemble models will be weak learners. Ensemble learning was a simple and powerful tool for turning these weak learners into strong learners.

Another insight from my models is that a neural network is often not the best solution for many data science tasks. In most applications, designing a neural network is more work than tuning the parameters of a less complex model. In this case, it also did not perform as well as boosting.
